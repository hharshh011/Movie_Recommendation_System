```python
df = df[selected_columns]
df
```

```python
df["genres"] = df["genres"].apply(lambda x: [d["name"] for d in literal_eval(x)])
df["keywords"] = df["keywords"].apply(lambda x: [d["name"] for d in literal_eval(x)])
```

```python
df["cast"] = df["cast"].apply(lambda x: [d["name"] for d in literal_eval(x)][:3])
```

```python
df["crew"]= df["crew"].apply(lambda x: [d["name"] for d in  literal_eval(df["crew"][0]) if d["job"]=="Director"])
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df = df.drop("tagline", axis = 1)
```

```python
df["overview"] = df["overview"].apply(lambda x: x.split())
```

```python
def remove_space(text):
    return [x.replace(" ", "") for x in text]
```

```python
df["genres"] = df["genres"].apply(remove_space)
df["keywords"] = df["keywords"].apply(remove_space)
df["cast"] = df["cast"].apply(remove_space)
df["crew"] = df["crew"].apply(remove_space)
```

```python
df["tag"] = df["genres"]+df["keywords"]+df["overview"]+df["cast"]+df["crew"]
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
def clean_text(text):
    
    doc = nlp(text)
    clean_text  = []
    for token in doc:
        if not token.is_stop and not token.is_punct and not token.is_space:
            clean_text.append(token.lemma_.lower()) # lemmatization and converting to lower case
            
    return " ".join(clean_text)
```

```python
def clean_text2(text):
    
    doc = nlp(text)
    clean_text  = []
    for token in doc:
        if token.pos_ in ["ADJ","NOUN"]:
            if not token.is_stop and not token.is_punct and not token.is_space:
                clean_text.append(token.lemma_.lower()) # lemmatization and converting to lower case
            
    return " ".join(clean_text)
```

```python
clean_text2("I am very happy going to school. I am also sad. I am following someone. I am trying to do")
```

```python
df["tag"]= df["tag"].apply(clean_text)
```

## Vectorisation

```python
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector_array = TfidfVectorizer(max_features=2000).fit_transform(df["tag"])
```

```python
vector_array
```

```python
pd.DataFrame(vector_array.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
m_m_similarity = cosine_similarity(vector_array)
```

```python
m_m_similarity.shape
```

```python
df.columns
```

```python
similarity_df = pd.DataFrame(m_m_similarity, columns=df["original_title"], index=df["original_title"])
similarity_df
```

```python
similarity_df["The Dark Knight Rises"].sort_values(ascending=False)[1:15]
```

```python
similarity_df.index
```

```python
def recommend_movies(movie_name, count= 5):
    try:
        return similarity_df[movie_name].sort_values(ascending=False)[1:count].index.values.tolist()
    except KeyError:
        raise Exception("Movie Recommendations not found. Try with a different movie")
    
```

```python
recommend_movies("Spectre", count = 5)
```
